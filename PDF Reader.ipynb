```python
# import library

import PyPDF2
```

```python
# open pdf file in read binary mode
f=open(r"C:\Users\Nirbhay\Downloads\Python.pdf","rb")

# parsing the file to pdf file reader & creating a pdf_reader object
pdf_reader=PyPDF2.PdfFileReader(f)

# showing total number of pages in PDF file
print(pdf_reader.numPages)

```

```python
# extracting each page's content

for pages in range(pdf_reader.numPages):
    
    # leave a line for better view
    print("\n")
    
    # starting of page
    print("=====STARTING OF {} PAGE=====".format(pages+1))
    
    # extracting data and view it
    data=pdf_reader.getPage(pages)
    print(data.extractText())
    
    # ending of page
    print("=====END OF {} PAGE=====".format(pages + 1))
    
    # leave a line for better view
    print("\n")
```
